# Data Cleaning
<p>In this stage, I started to get rid of any data I didn't need and to clean up everything else. I renamed some columns to make them easier for me to recall and split up the date and time.</p>
<p>I split up the dates and times into smaller slices since factors such as the specific minute or second of the tweet would have little impact on the outcome. Neither did I think the day would have much effect. </p>
<p>I also removed the favorited column since, after looking further into the data, I found that most of the sources had favorited set to false for every tweet.</p>
<p>For hours, I converted the datatype into integers. This is because I wanted to group them to make putting them into graphs easier. Otherwise, I would end up with 24 individual graphs for every hour of the day when having individual groupings would mean more and be easier.</p>
<p>Then I used a calculation to find the length of each tweet and save it as it's own column</p>
<p>I wanted to use sentiment analysis so I found some code online using textblobs and fed it my dataframe to obtain the sentiment of each tweet as well as other factors. The polarity shows where the sentiment tends towards with minus numbers being more negative and positive numbers being more positive. Subjectivity shows how much is personal opinion and how much is factual evidence with a higher subjectivity being more personal opinion.</p>

````python
import pandas as pd

#Reading the data
df = pd.read_csv('data/Task2/raw_twitter_data.csv', delimiter=',')

#Renaming columns to make them easier to remember and use
df = df.rename(columns={"favorite_count": "favourites", "retweet_count": "retweets", "is_quote_status": "quote"})

#Split columns by different symbols
df[['date', 'time']] = df.created_at.str.split(" ", expand = True)
df[['time', 'garbage']] = df.time.str.split("+", expand = True)
df[['year', 'month', 'day']] = df.date.str.split("-", expand = True)
df[['hour', 'minute', 'second']] = df.time.str.split(":", expand = True)
#df['date']= pd.to_datetime(df['date'])

#Converting hour to int so I can mathmatically group the hours
df = df.astype({'hour':'int'})

#Dropping stuff I don't need.
df.drop(['created_at', 'garbage', 'minute', 'second', 'time', 'date', 'day','favorited'], inplace=True, axis=1)

#https://sparkbyexamples.com/pandas/pandas-split-column/ Accessed 09/02/23
#https://sparkbyexamples.com/pandas/pandas-convert-column-to-int/ Accessed 03/03/23
````

````python
#Find length of each tweet using a lambda function and create its own column
df["length"] = df["tweet"].apply(lambda n: len(n.split()))

#https://thecleverprogrammer.com/2022/06/22/count-number-of-words-in-a-column-using-python/ Accessed 09/02/23
````

````python
from textblob import TextBlob
def sentiment_analysis(tweet):
    def getSubjectivity(text):
        return TextBlob(text).sentiment.subjectivity
  
    #Create a function to get the polarity
    def getPolarity(text):
        return TextBlob(text).sentiment.polarity
  
    #Create two new columns ‘Subjectivity’ & ‘Polarity’
    tweet['subjectivity'] = tweet['tweet'].apply(getSubjectivity)
    tweet ['polarity'] = tweet['tweet'].apply(getPolarity)
    def getAnalysis(score):
        if score < 0:
            return 'Negative'
        elif score == 0:
            return 'Neutral'
        else:
            return 'Positive'
    tweet ['sentiment'] = tweet['polarity'].apply(getAnalysis )
    return tweet
#https://towardsdatascience.com/my-absolute-go-to-for-sentiment-analysis-textblob-3ac3a11d524 Accessed 05/03/23
````

## Changing Data Types
<p>Since the mentions and hashtags were in an unusable format, I made a function to turn them into booleans instead. Booleans were easier with the format given than trying to count how much of each there was and I felt the exact number wasn't likely to have much effect unless the algorithm could recognize different hashtags and people being mentioned because those would be the most likely biggest deciding factor in what influence it has. </p>
<p>Then, I also grouped the hours into morning, afternoon, evening, and night, creating a column just for these groupings because I would need the regular hours later. </p>

````python
#Check if string is above 2 (Empty hashtags and mentions have two brackets)
def Has_Hashtags_or_Mentions(values):
    if len(values) <= 2:
        return False
    else:
        return True

#Apply function to both hashtags and mentions to convert them into booleans to see whether or not they have any mentions or hashtags
df['hashtags'] = df.hashtags.apply(Has_Hashtags_or_Mentions)
df['mentions'] = df.mentions.apply(Has_Hashtags_or_Mentions)

df.sample(10)
````

````python
#Return what time of day it is depending on what the hour is
def Group_Hours(hour):
    if hour < 6:
        return 'night'
    elif hour < 12:
        return 'morning'
    elif hour < 18:
        return 'afternoon'
    else:
        return 'evening'

#Apply outcome to new column called hour_group
df['hour_group'] = df.hour.apply(Group_Hours)

df.head()
````

### Calculating Alpha and outputting regression model
<p>Next, I performed the ridge repression itself, using a value of 0.1 for the alpha as determined by testing variations of alpha and plotting the results.</p>

````python
# Perform ridge regression
# Load the Ridge model utility
from sklearn.linear_model import Ridge
# Inspect the effect of varying alpha values
def ridge_vary_alpha(alphas):
    '''
    Input: An array of alphas.
    Output: A DataFrame containing the coefficients of Ridge regressions from each alpha.
    '''
    # Create an empty data frame
    df2 = pd.DataFrame()
    # Create a column of feature names
    df2['FeatureName'] = predictors.columns
    # For each alpha value in the list of alpha values,
    for alpha in alphas:
    # Create a lasso regression with that alpha value,
        ridge = Ridge(alpha=alpha)
        # Fit the lasso regression
        ridge.fit(X_train, y_train)
        # Create a column name for that alpha value
        column_name = f'Alpha = {alpha}, r2 = {ridge.score(X_train, y_train):.3f}'
        # Create a column of coefficient values
        df2[column_name] = ridge.coef_
        # Return the datafram
    return df2
alphas = 10**np.arange(-2,8).astype('float')
df_alpha = ridge_vary_alpha(alphas)
df_alpha
````

````python
# Load the Lasso model utility
from sklearn.linear_model import Lasso
# Inspect the effect of varying alpha values
def lasso_vary_alpha(alphas):
    '''
    Input: An array of alphas.
    Output: A DataFrame containing the coefficients of Ridge regressions from each alpha.
    '''
    # Create an empty data frame
    df2 = pd.DataFrame()
    # Create a column of feature names
    df2['FeatureName'] = predictors.columns
    # For each alpha value in the list of alpha values,
    for alpha in alphas:
    # Create a lasso regression with that alpha value,
        ridge = Lasso(alpha=alpha)
        # Fit the lasso regression
        ridge.fit(X_train, y_train)
        # Create a column name for that alpha value
        column_name = f'Alpha = {alpha}, r2 = {ridge.score(X_train, y_train):.3f}'
        # Create a column of coefficient values
        df2[column_name] = ridge.coef_
        # Return the datafram
    return df2
alphas = 10**np.arange(-1,3,0.1)
df_alpha = lasso_vary_alpha(alphas)
df_alpha
````

````python
df_alpha.plot()
plt.ylabel(r'$\beta$')
plt.xlabel(r'$log_{10}(\alpha)$')
plt.title(r'Varying $\alpha$ in Lasso Regression')
plt.axhline(0, c='k', linestyle=':')
plt.show()
````

# Conclusion and Discussion
<p>The overall conclusion is that all the models gave similar values for MSE and RMSE which suggests that they're all mostly the same. Lasso takes the magnitude of coefficients and ridge takes the square. Since Lasso is capable of setting coefficiants to zero, this allows it to select variables which are best, usually giving it an advantage over ridge which has to use every variable. For MSE score, the models achieved about 1.869 for all sources in ridge and 0.721 for just one while lasso had 1.562 together and 0.796. Meanwhile, multiple linear regression had 1.789 and 1.131 respectively. These scores seem to imply that lasso was best for all the sources together while ridge was just slightly better at singular sources. Multiple linear regression sits in the middle, being better than ridge at all sources together but not better than lasso and worse than both other models with the singular source.</p>
<p>Another thing I found was that the model works better overall on individual sources compared to all of them together. Therefore, the client would be better picking singular Twitter users they want to predict rather than trying to create an overall model and, for this task, ridge might be the better fit but not by a lot.</p>
